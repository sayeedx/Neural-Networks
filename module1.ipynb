```python
x = tf.range(-5.0, 5.0, 0.25)
x
```

```python
res = tf.nn.sigmoid(x)
res
```

```python
plt.plot(x, res)
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
res_tanh = tf.nn.tanh(x)
plt.plot(x.numpy(), res_tanh.numpy())
plt.title('Tanh Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
res_relu = tf.nn.relu(x)
plt.plot(x.numpy(), res_relu.numpy())
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
alpha = 0.1
res_leaky_relu = tf.nn.leaky_relu(x, alpha=alpha)
plt.plot(x.numpy(), res_leaky_relu.numpy())
plt.title(f'Leaky ReLU Function (alpha={alpha})')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
x = tf.range(-3.0, 3.0, 0.1)

res_sigm = tf.nn.sigmoid(x)
res_tanh = tf.nn.tanh(x)
res_relu = tf.nn.relu(x)
alpha = 0.1
res_lrelu = tf.nn.leaky_relu(x, alpha=alpha)
plt.plot(x.numpy(), res_sigm.numpy(), label='Sigmoid')
plt.plot(x.numpy(), res_tanh.numpy(), label='Tanh')
plt.plot(x.numpy(), res_relu.numpy(), label='ReLU')
plt.plot(x.numpy(), res_lrelu.numpy(), label=f'Leaky ReLU (alpha={alpha})')
plt.title('Comparison of Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
z = tf.constant([1.0, 2.0, 3.0])
softmax = tf.nn.softmax(z)

# Display the probabilities
print("Input logits:", z.numpy())
print("Probabilities:", softmax.numpy())
```

```python
# Vary the first logit from -2 to 5
z1 = tf.range(-2.0, 5.0, 0.1)
z2 = tf.constant(2.0)
z3 = tf.constant(3.0)

# Compute probabilities for each value of z1
probabilities = []
for value in z1:
    logits = tf.stack([value, z2, z3])
    probs = tf.nn.softmax(logits)
    probabilities.append(probs.numpy())

probabilities = np.array(probabilities)

# Plot the probabilities of each class
plt.plot(z1, probabilities[:, 0], label='Class 1')
plt.plot(z1, probabilities[:, 1], label='Class 2')
plt.plot(z1, probabilities[:, 2], label='Class 3')
plt.title('Softmax Function Probabilities')
plt.xlabel('Value of z1')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
